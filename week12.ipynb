```python
import csv
f=open('seoul.csv')
data=csv.reader(f)
header=next(data)
max_temp=-999 #최고 기온
max_date=''
min_temp=999 #최저 기온
min_date=''
dif_temp=-999 #일교차
dif_date=''

for row in data:
    if row[-1]=='':
        row[-1]=-999 #누락일 경우 최고기온 -999로 저장
    row[-1]=float(row[-1])#문자열로 저장된 최고 기온 값 실수로 반환
    if max_temp<row[-1]:#만약 지금까지 최고 기온보다 더 높다면 업데이트
        max_date=row[0]
        max_temp=row[-1]
    if row[-2]=='':
        row[-2]=999
    row[-2]=float(row[-2])
    if row[-2]<min_temp:
        min_date=row[0]
        min_temp=row[-2]
    if row[-1]=='' or row[-2]=='':
        row[-1]=-999
        row[-2]=999
    row[-1]=float(row[-1])
    row[-2]=float(row[-2])
    if (dif_temp)<(row[-1]-row[-2]):
        dif_date=row[0]
        dif_temp=(row[-1]-row[-2])
        
f.close()


print('기상 관측 아래 서울의 최고 기온이 가장 높았던 날은',max_date+'로',max_temp,'도 였습니다.')
print('기상 관측 아래 서울의 최저 기온이 가장 낮았던 날은',min_date+'로',min_temp,'도 였습니다.')
print('기상 관측 아래 서울의 일교차가 가장 높았던 날은',dif_date+'로',dif_temp,'도 였습니다.')
```

```python
import csv
f=open('subway.csv')
data=csv.reader(f)
header=next(data)
lineList = []
data2 = []
for row in data:
    if(row[1] not in lineList):
        lineList.append(row[1])
    data2.append(row)
    
for subwayLine in lineList:
    print(subwayLine)
    arr=[]
    for lineName in data2:
        if(subwayLine==lineName[1]):
            subwayPassenger=int(lineName[4].replace(",",""))+int(lineName[5].replace(",",""))
            arr.append([lineName[3],subwayPassenger])
    arr.sort(key=lambda x:x[1])
    print(" 이용객이 가장 많은 역 순위")
    print("{0:s} : {1:d} 명".format(arr[-1][0],arr[-1][1]))
    print("{0:s} : {1:d} 명".format(arr[-2][0],arr[-2][1]))
    print("{0:s} : {1:d} 명".format(arr[-3][0],arr[-3][1]))
    print(" 이용객이 가장 적은 역 순위")
    print("{0:s} : {1:d} 명".format(arr[0][0],arr[0][1]))
    print("{0:s} : {1:d} 명".format(arr[1][0],arr[1][1]))
    print("{0:s} : {1:d} 명".format(arr[2][0],arr[2][1]))
f.close()
```
